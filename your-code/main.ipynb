Additional info: Over a century ago, an Italian economist named Vilfredo Pareto discovered that roughly 20% of the customers account for 80% of the typical retail sales. This is called the [Pareto principle](https://en.wikipedia.org/wiki/Pareto_principle). Check if this dataset displays this characteristic.

```python
#check for non integers
print(cust_df.dtypes)
print(cust_df.head())

#check for missing values
print(cust_df.isnull().sum())

#check for correlation between columns
corrolation_matrix = cust_df.corr(numeric_only=True)

#plots correlational heatmap
plt.figure(figsize=[10, 6])
sns.heatmap(corrolation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
#check df through pareto priniciple

print(cust_df.columns)

# creates a column that totals the sum of each customers spending
cust_df["Total_Spending"] = cust_df[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].sum(axis=1)

# Sort by total spending
sorted_spending = cust_df["Total_Spending"].sort_values(ascending=False).reset_index(drop=True)

# Cumulative spending
cumulative_spending = sorted_spending.cumsum()
total_spending = sorted_spending.sum()
cumulative_percent = cumulative_spending / total_spending

# Find how many customers account for 80% of spending
num_customers = len(cust_df)
threshold_80 = cumulative_percent[cumulative_percent <= 0.8]
pareto_cutoff = len(threshold_80) / num_customers * 100

print(f"Top {pareto_cutoff:.2f}% of customers account for 80% of the total spending.")

# Plotting Pareto curve
plt.figure(figsize=(10, 6))
plt.plot(cumulative_percent.values * 100, label="Cumulative % of Spending")
plt.axhline(y=80, color='r', linestyle='--', label="80% Line")
plt.axvline(x=len(threshold_80), color='g', linestyle='--', label=f"{pareto_cutoff:.1f}% of Customers")
plt.title("Pareto Principle in Wholesale Customers Data")
plt.xlabel("Customers (sorted by spending)")
plt.ylabel("Cumulative % of Total Spending")
plt.legend()
plt.grid(True)
plt.show()
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

#normalizes the data so that numerical outliers don't dominate the learning process and numeric features are brought into a standard range.
scaler = StandardScaler()

X_scaled = scaler.fit_transform(cust_df)


# Initialize the K-Means model
kmeans = KMeans(n_clusters=3, random_state=42)  # change `n_clusters` as needed

# Fit the model to the data
kmeans.fit(X_scaled)

# Assign cluster labels to original DataFrame
cust_df["Label"] = kmeans.labels_

# Preview the clustered data
print(cust_df.head())

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal k')
plt.grid(True)
plt.show()
```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
kmeans_2 = KMeans(n_clusters=2).fit(customers_scale)
print(kmeans_2)

labels = kmeans_2.predict(customers_scale)
print(labels)

clusters = kmeans_2.labels_.tolist()
print(clusters)
```

```python
cust_df['Label'] = clusters
print(cust_df)
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN 

# Your code here
dbscan = DBSCAN(eps=0.5, min_samples=2) 
dbscan.fit(X_scaled)

# Step 3: Add cluster labels back to the original dataframe
cust_df['labels_DBSCAN'] = dbscan.labels_  # -1 indicates noise (outliers)

cust_df.head()
```
